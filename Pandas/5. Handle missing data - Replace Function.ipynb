# If data has unit-of-measure 

```python
import pandas as pd
import numpy as np
df = pd.read_csv("sheet5-weather_data.csv")
df
```

```python
new_df=df.replace('[A-Za-z]',' ', regex=True)
new_df
```

# Replacing a list of values with another list of values

```python
df=pd.DataFrame({
    'score' : ['exceptional','average','good','poor','average','exceptional'],
    'student' : ['rob', 'maya', 'arfa', 'sanu', 'maddu', 'ram']
})
df
```

```python
df.replace(['poor','average','good','exceptional'], [1,2,3,4])
```
