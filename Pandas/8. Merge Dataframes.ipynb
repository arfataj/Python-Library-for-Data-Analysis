```python
import pandas as pd
df1=pd.DataFrame({
    "city":["new york","chicago","orlando","san francisco"],
    "temperature": [21,14,12,67]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","orlando","baltimore"],
    "humidity": [65,67,75,67]
})
df2
```

```python
df3=pd.merge(df1,df2, on="city")
df3
```

# Type Of DataBase Joins

```python
df3=pd.merge(df1,df2, on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2, on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df2, on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2, on="city",how="right")
df3
```

# indicator flag

```python
df3=pd.merge(df1,df2,on="city", how="outer", indicator=True)
df3
```

# suffixes

```python
df1=pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimer"],
    "temperature":[21,45,23,67],
    "humidity": [65,78,34,78]
})
df1
```

```python
df2=pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature":[12,23,67],
    "humidity": [65,78,67]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df3=pd.merge(df1,df2,on="city",suffixes=('_left','_right'))
df3
```

# join

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index('city',inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2,lsuffix='_l', rsuffix='_r')
```
